# 1. Split the dataset into two subsets as follows:

### Subset 1: the applicants with Checking = 1 or Checking = 2

### Subset 2: the applicants where Checking = 3 or Checking = 4
    
Clean the subsets if necessary.


```python
subset1 = data[(data['Checking'] == 1) | (data['Checking'] == 2)]
subset2 = data[(data['Checking'] == 3) | (data['Checking'] == 4)]
subset1
```

# 2. For each subset, establish a training set and validation set. 

```python
X_subset1 = subset1.drop('Checking', axis=1)
y_subset1 = subset1['Checking']
X_train_subset1, X_val_subset1, y_train_subset1, y_val_subset1 = train_test_split(
    X_subset1, y_subset1, test_size=0.2, stratify=y_subset1, random_state=42)

X_subset2 = subset2.drop('Checking', axis=1)
y_subset2 = subset2['Checking']
X_train_subset2, X_val_subset2, y_train_subset2, y_val_subset2 = train_test_split(
    X_subset2, y_subset2, test_size=0.2, stratify=y_subset2, random_state=42)
```

X_train_subset1, y_train_subset1: Training features and target for Subset 1.

X_val_subset1, y_val_subset1: Validation features and target for Subset 1.

X_train_subset2, y_train_subset2: Training features and target for Subset 2.

X_val_subset2, y_val_subset2: Validation features and target for Subset 2.

test_size=0.2 to allocate 20% of the data to the validation set and 80% to the training set.
